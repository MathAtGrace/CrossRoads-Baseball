```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://gclancers.com/sports/baseball/stats'
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
data = soup.find('table', class_='sidearm-table')
data
```

```python
headers = data.find('thead', class_='sidearm-primary')
headers
```

```python
headers_iterable = headers.find_all('th')
headers_iterable
```

```python
table_data = [header.text for header in headers_iterable]
print(table_data)
```

```python
df = pd.DataFrame(columns = table_data)
df
```

```python
content = data.find('tbody')
content
```

```python
for span in content.find_all('span'):
    span.decompose()
```

```python
table_rows = content.find_all('tr')
table_rows
```

```python
for row in table_rows:
    table_content = row.find_all(['td', 'th'])
    individual_data = [data.text.strip() for data in table_content]
    df.loc[len(df)] = individual_data
df
```

```python
df.to_csv('baseball_data.csv', index=False)
```
