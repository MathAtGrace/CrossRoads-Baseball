```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
urls = [
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/taylorin?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/gracein?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/mountvernonnazareneoh?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/indianawesleyanin?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/huntingtonin?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/springarbormi?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/saintfrancisin?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/bethelin?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/marianin?view=lineup',
    'https://crossroadsleague.com/sports/bsb/2024-25/teams/goshenin?view=lineup'
]
```

```python
driver = webdriver.Firefox()
```

```python
all_data = []
```

```python
for url in urls:
    driver.get(url)
    time.sleep(2)

    page = driver.page_source
    with open('../raw-data/baseball.html', 'w') as file:
        file.write(page)
    
    soup = BeautifulSoup(page, 'html.parser')

    lineup = soup.find('div', id='lineup')

    headers = [header.text.strip() for header in lineup.find('thead').find_all('th')]

    rows = lineup.find('tbody').find_all('tr')
    for row in rows:
        individual_data = [data.text.strip() for data in row.find_all(['td', 'th'])]

        if len(individual_data) == len(headers):
            row_data = dict(zip(headers, individual_data))
            row_data["team"] = soup.find('div', class_='team-attributes').find_all('span', class_='d-block')[1].text
            all_data.append(row_data)
```

```python
driver.quit()
```

```python
df = pd.DataFrame(all_data)
df
```

```python
df.to_csv('../data/baseball_data.csv', index=False)
```
